Question 1 using Breadth First search

```python
import copy
q = []
visited = []
def up(s):
    temp = copy.deepcopy(s)
    pos = find_pos(temp)
    i = pos[0]
    j = pos[1]

    if i>0:
        temp[i][j]=temp[i-1][j]
        temp[i-1][j] = -1
    return temp

def down(s):
    temp = copy.deepcopy(s)
    pos = find_pos(temp)
    i = pos[0]
    j = pos[1]

    if i<2:
        temp[i][j]=temp[i+1][j]
        temp[i+1][j] = -1
    return temp

def right(s):
    temp = copy.deepcopy(s)
    pos = find_pos(temp)
    i = pos[0]
    j = pos[1]

    if j<2:
        temp[i][j]=temp[i][j+1]
        temp[i][j+1] = -1
    return temp

def left(s):
    temp = copy.deepcopy(s)
    pos = find_pos(temp)
    i = pos[0]
    j = pos[1]

    if j>0:
        temp[i][j]=temp[i][j-1]
        temp[i][j-1] = -1
    return temp

def find_pos(s):
    for i in range(len(s)):
        for j in range(len(s[0])):
            if s[i][j]==-1:
                return ([i,j])

def compare(s,g):
    for i in range(len(s)):
        for j in range(len(s[i])):
            if s[i][j]!=g[i][j]:
                return (0)
    return (1)
def enqueue(new):
    global q
    q+=[new]

def dequeue():
    global q
    new = q[0]
    del q[0]
    return(new)


def search(s,g):

    global q
    global visited
    visited+=[s]
    if compare(s,g):
        print("found")
        print(len(visited))
        exit()
    while(1):
        new = up(s)
        if new==g:
            print("found")
            print(len(visited))
            exit()
        else:
            if new not in visited and new not in q:
                enqueue(new)
                visited+=[new]


        new = down(s)
        if new==g:
            print("found")
            print(new)
            print(len(visited))
            exit()
        else:
            #q+=[new]
            if new not in visited and new not in q:
                enqueue(new)
                visited+=[new]


        new = right(s)
        if new==g:
            print("found")
            print(len(visited))
            exit()
        else:
            # q+=[new]
            if new not in visited and new not in q:
                enqueue(new)
                visited+=[new]


        new = left(s)
        if new==g:
            print("found")
            print(len(visited))
            exit()
        else:
            # q+=[new]
            if new not in visited and new not in q:
                enqueue(new)
                visited+=[new]
        if len(q)>0:
            # s = q[0]
            # del q[0]
            temp = dequeue()
            s = temp
        else:
            print("not possible")


def main():
    start = [[1,2,3],
            [8,-1,4],
            [7,6,5]]
    goal = [[2,8,1],
            [-1,4,3],
            [7,6,5]]

    ans = compare(start,goal)
    print(start)
    search (start,goal)


if __name__ == "__main__":
    main()
```

```python
def Next(toCap, fromJCap, d):

    fromJ = fromJCap
    to = 0

    step = 1
    while ((fromJ is not d) and (to is not d)):

        temp = min(fromJ, toCap-to)

        to = to + temp
        fromJ = fromJ - temp

        step = step + 1
        if ((fromJ == d) or (to == d)):
            break

        if fromJ == 0:
            fromJ = fromJCap
            step = step + 1

        if to == toCap:
            to = 0
            step = step + 1

    return step

def gcd(a, b):
    if b==0:
        return a
    return gcd(b, a%b)

def min(n, m, d):
    if m> n:
        temp = m
        m = n
        n = temp

    if (d%(gcd(n,m)) != 0):
        return -1

    return(min(Next(n,m,d), Next(m,n,d)))

def main():
    n = 4
    m = 3
    d = 2

    print('output at ', min(n, m, d))

if __name__ == "__main__":
    main()
```

```python
from sys import maxsize
from itertools import permutations
V = 4

def sales(graph, s):

  vertex = []
  for i in range(V):
    if i != s:
      vertex.append(i)

  minpath = maxsize
  nextperm=permutations(vertex)
  for i in nextperm:

    cr= 0

    k = s
    for j in i:
      cr+= graph[k][j]
      k = j
    cr+= graph[k][s]

    minpath = min(minpath, cr)

  return minpath

if __name__ == "__main__":


  graph = [[0, 10, 15, 20], [10, 0, 35, 25],
      [15, 35, 0, 30], [20, 25, 30, 0]]
  s = 0
  print(sales(graph, s))
```
